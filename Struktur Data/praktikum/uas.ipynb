```python
# soal 1
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def level_order(temp):
    if not temp:
        return
    q = []
    q.append(temp)
    while q:
        temp = q.pop(0)
        print(temp.data, end=" ")
        if temp.left:
            q.append(temp.left)
        if temp.right:
            q.append(temp.right)

def delDeep(root, d_node):
    q = []
    q.append(root)
    while len(q):
        temp = q.pop(0)
        if temp is d_node:
            temp = None
            return
        if temp.right:
            if temp.right is d_node:
                temp.right = None
                return
            else:
                q.append(temp.right)
        if temp.left:
            if temp.left is d_node:
                temp.left = None
                return
            else:
                q.append(temp.left)

def delet(root, key):
    if root == None:
        return None
    if root.left == None and root.right == None:
        if root.data == key:
            return None
        else:
            return root
    k_node = None
    q = []
    q.append(root)
    temp = None
    while len(q):
        temp = q.pop(0)
        if temp.data == key:
            k_node = temp
        if temp.left:
            q.append(temp.left)
        if temp.right:
            q.append(temp.right)
    if k_node:
        k_node.data = temp.data
        delDeep(root, temp)
    return root

def insert(root, data):
    q = []
    q.append(root)
    while len(q):
        temp = q.pop(0)
        if not temp.left:
            temp.left = Node(data)
            break
        else:
            q.append(temp.left)
        if not temp.right:
            temp.right = Node(data)
            break
        else:
            q.append(temp.right)

root = Node(8)
root.left = Node(3)
root.right = Node(15)
root.right.left = Node(12)
root.right.right = Node(7)

print("sebelum dihapus")
level_order(root)

root = delet(root, 12)
print("\nsetelah dihapus:")
level_order(root)

insert(root, 5)
insert(root, 6)
print("\nsetelah diinsert:")
level_order(root)
```

```python
#soal 2
class Node:
    def __init__(self, key, value):
        self.key = key
        self.value = value
        self.next = None

class separatechaining:
    def __init__(self, size):
        self.size = size
        self.table = [None] * size;

    def hash_function(self, key):
        return hash(key) % self.size;

    def insert(self, key, value):
        index = self.hash_function(key)
        if self.table[index] is None:
            self.table[index] = Node(key, value)
        else:
            current = self.table[index]
            while current.next:
                current = current.next
            current.next = Node(key, value)

    def search(self, key):
        index = self.hash_function(key)
        current = self.table[index]
        while current:
            if current.key == key:
                return current.value
            current = current.next
        return None

hash_table = separatechaining(10)
hash_table.insert("Paku", 5)
hash_table.insert("Palu", 10)
hash_table.insert("Gergaji", 15)
hash_table.insert("Skop", 20)
hash_table.insert("Pasah", 25)

print(hash_table.search("Skop"))
print(hash_table.search("Pasah"))
print(hash_table.search("Palu"))
```

```python
# soal 3
hutang = {
    'Ani': {'Budi': 5000, 'Doni': 10000},
    'Budi': {'Cahyo': 10000},
    'Cahyo': {'Ani': 1000, 'Eko': 10000},
    'Doni': {'Cahyo': 5000},
    'Eko': {'Budi': 5000, 'Doni': 500}
}

# Menghitung saldo bersih untuk setiap orang
saldo = {}
for orang in hutang:
    for kreditor, jumlah in hutang[orang].items():
        saldo[orang] = saldo.get(orang, 0) - jumlah
        saldo[kreditor] = saldo.get(kreditor, 0) + jumlah

debitor = [(orang, jumlah) for orang, jumlah in saldo.items() if jumlah < 0]
kreditor = [(orang, jumlah) for orang, jumlah in saldo.items() if jumlah > 0]

# Mengurutkan debitor dan kreditor dari jmlh hutang 
debitor.sort(key=lambda x: x[1])
kreditor.sort(key=lambda x: x[1], reverse=True)

transaksi = []

# Menyelesaikan hutang
i, j = 0, 0
while i < len(debitor) and j < len(kreditor):
    orang_hutang, jumlah_hutang = debitor[i]
    orang_piutang, jumlah_piutang = kreditor[j]
    
    jumlah_setel = min(-jumlah_hutang, jumlah_piutang)
    transaksi.append(f"{orang_hutang} membayar {orang_piutang} {jumlah_setel}")
    
    debitor[i] = (orang_hutang, jumlah_hutang + jumlah_setel)
    kreditor[j] = (orang_piutang, jumlah_piutang - jumlah_setel)
    
    if debitor[i][1] == 0:
        i += 1
    if kreditor[j][1] == 0:
        j += 1

for trans in transaksi:
    print(trans)
```
