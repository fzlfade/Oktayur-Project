```python
# Fungsi untuk melakukan merge sort (Soal nomor 1)
def merge_sort(siswa):
    if len(siswa) <= 1:
        return siswa

    mid = len(siswa) // 2
    left = merge_sort(siswa[:mid])
    right = merge_sort(siswa[mid:])

    return merge(left, right)

# Menggabungkan dua list yang sudah diurutkan (Soal nomor 1)
def merge(left, right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result

# Memanggil siswa berdasarkan nomor induk yang sudah diurutkan (Soal nomor 1)
def panggil_siswa(siswa):
    siswa = merge_sort(siswa)

    print("(Nomor 1 mengurutkan siswa menggunakan sort merge)")
    for current_siswa in siswa:
        print(f"Siswa dengan nomor induk {current_siswa} dipanggil untuk wawancara.")

# Mengumumkan hasil ujian menggunakan linier search (Soal nomor 2)
def umumkan_hasil(hasil_ujian):
    siswa_ulang = []
    siswa_sempurna = []

    for nama, nilai in hasil_ujian.items():
        if nilai < 60:
            siswa_ulang.append(nama) # Jika nilai siswa dibawah 60 maka masukan kedalam list siswa_ulang
        elif nilai == 100:
            siswa_sempurna.append(nama) # Jika nilai siswa 100 maka masukan kedalam list siswa_sempurna
    print("\n(Nomor 2 mengumumkan hasil ujian siswa menggunakan linier search)")

    print("Siswa yang harus mengulang minggu depan:")
    for siswa in siswa_ulang:
        print(siswa, "dengan nilai", hasil_ujian[siswa])

    print("\nSiswa yang mendapatkan nilai sempurna:")
    for siswa in siswa_sempurna:
        print(siswa, "dengan nilai", hasil_ujian[siswa]) 



# Menjalankan program
siswa = [2320506044, 2320506042, 2320506041, 2320506043, 2320506040] # Daftar siswa berdasarkan nomor induk (soal nomor 1)
panggil_siswa(siswa) # Soal nomor 1

# Hasil ujian siswa (Soal nomor 2)
hasil_ujian = {
    "Oktario Mufti Yudha": 95,
    "Ilham Monitor": 55,
    "Faizal Avatar": 100,
    "Sunny Laprak": 70,
    "Restu Gaming": 50,
    "Iwan Caleg": 45
}
umumkan_hasil(hasil_ujian)  # Memanggil fungsi (soal nomor 2)
```
