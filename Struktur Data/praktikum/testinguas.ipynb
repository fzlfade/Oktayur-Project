```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def inorder(temp):
    if(not temp):
        return
    inorder(temp.left)
    print(temp.data, end=" ")
    inorder(temp.right)

def delDeep(root, d_node):
    q = []
    q.append(root)
    while(len(q)):
        temp = q.pop(0)
        if temp is d_node:
            temp = None
            return
        if temp.right:
            if temp.right is d_node:
                temp.right = None
                return
            else:
                q.append(temp.right)
        if temp.left:
            if temp.left is d_node:
                temp.left = None
                return
            else:
                q.append(temp.left)

def delet(root, key):
    if root == None:
        return None
    if root.left == None and root.right == None:
        if root.data == key:
            return None
        else:
            return root
    k_node = None
    q = []
    q.append(root)
    temp = None
    while(len(q)):
        temp = q.pop(0)
        if temp.data == key:
            k_node = temp
        if temp.left:
            q.append(temp.left)
        if temp.right:
            q.append(temp.right)
    if k_node:
        k_node.data = temp.data
        delDeep(root, temp)
    return root

def insert(root, data):
    q = []
    q.append(root)
    while(len(q)):
        temp = q.pop(0)
        if not temp.left:
            temp.left = Node(data)
            break
        else:
            q.append(temp.left)
        if not temp.right:
            temp.right = Node(data)
            break
        else:
            q.append(temp.right)

root = Node(8)
root.left = Node(3)
root.right = Node(15)
root.right.left = Node(12)
root.right.right = Node(7)

print("Tree sebelum proses hapus:")
inorder(root)


root = delet(root, 12)
print("\nTree setelah proses hapus:")
inorder(root)

insert(root, 5)
insert(root, 6)
print("\nTree setelah proses insert:")
inorder(root)
```

```python
class Node:
    def __init__(self, key, value):
        self.key = key
        self.value = value
        self.next = None

class separatechaining:
    def __init__(self, size):
        self.size = size
        self.table = [None] * size;

    def hash_function(self, key):
        return hash(key) % self.size;

    def insert(self, key, value):
        index = self.hash_function(key)
        if self.table[index] is None:
            self.table[index] = Node(key, value)
        else:
            current = self.table[index]
            while current.next:
                current = current.next
            current.next = Node(key, value)

    def search(self, key):
        index = self.hash_function(key)
        current = self.table[index]
        while current:
            if current.key == key:
                return current.value
            current = current.next
        return None

hash_table = separatechaining(10)
hash_table.insert("Paku", 5)
hash_table.insert("Palu", 10)
hash_table.insert("Gergaji", 15)
hash_table.insert("Skop", 20)
hash_table.insert("Pasah", 25)

print(hash_table.search("Skop"))
print(hash_table.search("Pasah"))
print(hash_table.search("Palu"))
```

```python
def min_transaksi(hutang):
    # Menghitung total hutang dan piutang untuk setiap orang
    saldo = {}
    for penghutang, pemberi_pinjaman, jumlah in hutang:
        saldo[penghutang] = saldo.get(penghutang, 0) - jumlah
        saldo[pemberi_pinjaman] = saldo.get(pemberi_pinjaman, 0) + jumlah

    # Membuat daftar hutang dan piutang
    daftar_penghutang = sorted((jumlah, orang) for orang, jumlah in saldo.items() if jumlah < 0)
    daftar_pemberi_pinjaman = sorted((jumlah, orang) for orang, jumlah in saldo.items() if jumlah > 0)

    transaksi = []

    # Membayar hutang dengan piutang
    while daftar_penghutang and daftar_pemberi_pinjaman:
        hutang, penghutang = daftar_penghutang.pop()
        piutang, pemberi_pinjaman = daftar_pemberi_pinjaman.pop()

        # Jumlah yang dibayar adalah minimum antara hutang dan piutang
        pembayaran = min(-hutang, piutang)

        transaksi.append((penghutang, pemberi_pinjaman, pembayaran))

        # Memperbarui hutang dan piutang
        hutang += pembayaran
        piutang -= pembayaran

        if hutang < 0:
            daftar_penghutang.append((hutang, penghutang))
        if piutang > 0:
            daftar_pemberi_pinjaman.append((piutang, pemberi_pinjaman))

    return transaksi

hutang = [
    ("Ani", "Budi", 5000),
    ("Ani", "Doni", 10000),
    ("Budi", "Cahyo", 10000),
    ("Cahyo", "Ani", 1000),
    ("Cahyo", "Eko", 10000),
    ("Doni", "Cahyo", 5000),
    ("Eko", "Budi", 5000),
    ("Eko", "Doni", 500),
]

print(min_transaksi(hutang))
```
