```python
import pandas as pd

# Load dataset dari file .csv dan ambil 5 data teratas
df = pd.read_csv('dataset-tanaman-padi.csv')
datasetTeratas = df.head(5)

print('5 Dataset sebelum pembersihan:\n', datasetTeratas)

# Mereplace data yang tidak memiliki kelembapan dengan nilai tengah kelembapan ke 5 dataset teratas
datasetTeratas['Kelembapan'].fillna(datasetTeratas['Kelembapan'].median(), inplace=True)
# Mereplace data yang tidak memiliki Curah hujan dengan rata rata Curah hujan ke 5 dataset teratas
datasetTeratas['Curah hujan'].fillna(datasetTeratas['Curah hujan'].mean(), inplace=True)
# Menghapus data yang tidak memiliki nama Provinsi
datasetTeratas.dropna(subset=['Provinsi'], inplace=True)

print('setelah pembersihan:\n', datasetTeratas)
```

### Pertemuan 13 ###

```python
import matplotlib.pyplot as plt

kategori = ['Suhu rata-rata', 'Kelembapan']
nilai = [datasetTeratas['Suhu rata-rata'].mean(), datasetTeratas['Kelembapan'].mean()]

plt.figure(figsize=(5, 3))
plt.bar(kategori, nilai, color='orange')
plt.xlabel('Kategori')
plt.ylabel('Nilai')
plt.title('Bar Chart kategori pada Provinsi Aceh')
plt.show()
```

```python
import seaborn as sns
import pandas as p

# Plot
sns.regplot(x='Curah hujan', y='Suhu rata-rata', data=datasetTeratas, color='orange')
plt.title('Hubungan antara Curah hujan dan suhu rata rata')
plt.figure(figsize=(5, 3))
plt.show()
```
